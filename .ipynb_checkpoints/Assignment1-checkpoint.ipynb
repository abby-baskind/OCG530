## 1) Mixing Ratios - Two Sources (20 points)

This problem uses the fact that $\Theta$ and $S_A$ are conserved to learn about the composition of water found in the ocean interior.

Imagine you have two "source" water masses: a polar one ($\Theta_1 = 2^\circ$ C, $S_1 = 32$ g/kg) and a tropical one ($\Theta_1 = 20^\circ$ C, $S_1 = 35$ g/kg).
These two water masses mix in the interior to form thermocline water.

1. What is the $\Theta$ and $S$ of the thermocline water assuming that each source water mass makes a 50/50 contribution?
2. What if the polar water only contributes 10% and the tropical water contributes 90%?
3. Find the general rule for the thermocline water composition given the mixing ratio $R$. ($R=1$ means 100% polar water; $R=0$ means 100% tropical water)

### 1.1) What is the $\Theta$ and $S$ of the thermocline water assuming that each source water mass makes a 50/50 contribution?

* mixing is linear
* $V_{total}C_{total} = V_1 C_1 + V_2 C_2$
* $V_{total} = V_1 + V_2$

```python
# Water masses
theta1 = 2
S1 = 32
theta2 = 20
S2 = 35

# Linear Mixing
Vtotal = 1
V1 = 0.5
V2 = 0.5
Stotal = (V1*S1 + V2*S2)/Vtotal
thetatotal = (V1*theta1 + V2*theta2)/Vtotal

# Answers
S_total = "%.3f" % Stotal
stri = "S = " + S_total + ' g/kg'
print(stri)
theta_total = "%.3f" % thetatotal
stri = "T = " + theta_total + ' 째C'
print(stri)
```

### 1.2) What if the polar water only contributes 10% and the tropical water contributes 90%?

```python
# Water masses
theta1 = 2
S1 = 32
theta2 = 20
S2 = 35

# Linear Mixing
Vtotal = 1
V1 = 0.1
V2 = Vtotal - V1
Stotal = (V1*S1 + V2*S2)/Vtotal
thetatotal = (V1*theta1 + V2*theta2)/Vtotal

# Answers
S_total = "%.3f" % Stotal
stri = "S = " + S_total + ' g/kg'
print(stri)
theta_total = "%.3f" % thetatotal
stri = "T = " + theta_total + ' 째C'
print(stri)
```

### 1.3) Find the general rule for the thermocline water composition given the mixing ratio $R$. ($R=1$ means 100% polar water; $R=0$ means 100% tropical water)

$C_{total} = R C_1 + (1 - R) C_2 = RC_1 + C_2 - RC_2$

```python
# Water masses
theta1 = 2
S1 = 32
theta2 = 20
S2 = 35

# Linear Mixing
Vtotal = 1
V1 = np.linspace(0,1)
V2 = Vtotal - V1
Stotal = (V1*S1 + V2*S2)/Vtotal
thetatotal = (V1*theta1 + V2*theta2)/Vtotal

fig, axarr = plt.subplots(ncols = 2, figsize = [10,5])
axarr = axarr.flatten()

ax = axarr[0]
ax.plot(V1,Stotal)
ax.scatter(0,35, label = "Tropical Water", color = 'red')
ax.scatter(1,32, label = "Polar Water", color = 'blue')
ax.set_xlabel('Fraction of Polar Water')
ax.set_ylabel('Salinity')
ax.legend()

ax = axarr[1]
ax.plot(V1,thetatotal)
ax.scatter(0,20, label = "Tropical Water", color = 'red')
ax.scatter(1,2, label = "Polar Water", color = 'blue')
ax.set_xlabel('Fraction of Polar Water')
ax.set_ylabel('Temperature')
ax.legend()
```

### 3.3) Comment on the role of the thermobaric effect in the global ocean stratification.

#### An Approximate Equation of State ##

This equation contains the "essential" nonlinearities in the equation of state of seawater.

$$ \rho \simeq \rho_0 \left [ 1 - \frac{gz}{c_0^2} - \alpha_0(1 - \gamma_B z)(\Theta - \Theta_0)
- \frac{\gamma_C}{2}(\Theta - \Theta_0)^2 + \beta_0 (S_A - S_{A0} ) \right ]
$$

where $\rho_0$, $c_0$, $\alpha_0$, $\gamma_B$, $\gamma_C$, and $\beta_0$ are all positive constants.
Here we have assumed that pressure $p$ is simply proportional to depth $z$. (As we will see next, this is a very good approximation.)

The parameter $\gamma_B$ expresses the nonlinear effect of __thermobaricity__. The parameter $\gamma_C$ expresses the nonlinear effect of __cabbeling__.

```python
rho0 = 1027
g = 9.81
c = 1500
alpha = 1.67e-4
gammaB = 1.12e-4
gammaC = 1e-5
beta = 0.78e-3
T0 = 10
S0 = 35
T = np.linspace(0,20)
S = 35
z = np.linspace(-2000,0)
[T,z] = np.meshgrid(T,z)

# Density with thermobaric effect
rho_B = rho0 * (1 - (g*z)/(c*c) - alpha * (1 - gammaB*z) * (T - T0) - (gammaC/2) * (T - T0)**2 + beta * (S - S0))

# Density with no thermobaric effect (gammaB = 0)
rho = rho0 * (1 - (g*z)/(c*c) - alpha * (1 - 0*z) * (T - T0) - (gammaC/2) * (T - T0)**2 + beta * (S - S0))

# Difference between thermobaric density and non-thermbaric density
drho = rho_B - rho

fig, ax = plt.subplots()
im = ax.pcolormesh(T,z,drho, cmap = 'coolwarm')
cbar = plt.colorbar(im, ax = ax)
cbar.set_label(r'$\Delta \rho = \rho_{\gamma_B > 0} - \rho_{\gamma_B = 0}$')
ax.set_xlabel('Temperature 째C')
ax.set_ylabel('Depth [m]')
ax.axvline(10, color = 'k', label = 'Initial T')
ax.legend()
```

```python
rho0 = 1027
g = 9.81
c = 1500
alpha = 1.67e-4
gammaB = 1.12e-4
gammaC = 1e-5
beta = 0.78e-3
T0 = 10
S0 = 35
S = 35
z = np.linspace(-2000,0)

def drho_b(T):
    # Density with thermobaric effect
    rho_B = rho0 * (1 - (g*z)/(c*c) - alpha * (1 - gammaB*z) * (T - T0) - (gammaC/2) * (T - T0)**2 + beta * (S - S0))
    # Density with no thermobaric effect (gammaB = 0)
    rho = rho0 * (1 - (g*z)/(c*c) - alpha * (1 - 0*z) * (T - T0) - (gammaC/2) * (T - T0)**2 + beta * (S - S0))
    # Difference between thermobaric density and non-thermbaric density
    drho = rho_B - rho
    return drho

fig, ax = plt.subplots()
ax.plot(drho_b(0),z, label = 'T = 0', color = 'blue')
ax.plot(drho_b(5),z, label = 'T = 5', color = 'cyan')
ax.plot(drho_b(15),z, label = 'T = 15', color = 'hotpink')
ax.plot(drho_b(20),z, label = 'T = 20', color = 'red')

ax.set_xlabel(r'$\Delta \rho = \rho_{\gamma_B > 0} - \rho_{\gamma_B = 0}$')
ax.set_ylabel('Depth [m]')
ax.legend()
```

Thermobaricity, when the in situ temperature is greater than the reference temperature, works to decrease density with depth. When in situ temperature is less than the reference temperature, thermobaricity works to increase density with depth. Overall, the effect of thermobaricity increases with depth.

Considering the deep ocean is generally cooler than overlying water, we can assume (under these conditions) that thermobaricity generally increases the density of the cooler, underlying water. This is to say, pressure increases the density of deep/bottom water to reinforce stratification.

### 4.1) How long would it take to evaporate away the entire basin if there were no balancing processes? (This has actually happened in the geological past during the [Messian Salinity Crisis](https://en.wikipedia.org/wiki/Messinian_salinity_crisis). Check it out!)

```python
Amed = 2500000 * 1e3 * 1e3 # m2
Hmed = 1500
Vmed = Amed * Hmed
V = "%.4f" % Vmed
stri = 'Volume of Med Sea = ' + V + ' m3'
print(stri)
```

```python
EP = 0.8 # m/yr
EP_A = EP * Amed # m3/yr
TotEP = "%.4f" % EP_A
stri = 'Total Evap Minus Precip (Area Integrated) = ' + TotEP + ' m3/yr'
print(stri)
```

```python
# Vfinal = Vinitial - EP * years
# EP * years = Vinitial - Vfinal
# years = dV/EP
years = Vmed/EP_A
year = "%.0f" % years
stri = 'Years to evaporate Med Sea = ' + year + ' years'
print(stri)
```

### 4.2) Sea level change in the Mediterranean is measured in mm year$^{-1}$ (that is, a factor of 100 smaller than the evaporation minus precipitation imbalance).  Therefore, there must be a supply of volume into the Mediterranean that approximately balances the net evaporation. Give that volume supply in Sverdrups.

1 Sv = 10$^6$ m$^3$/s

$$
A_{MedSea}*H = A_{MedSea}*H_0 - EP*A_{MedSea}*t + \text{Supply} * \frac{3153600\ s}{1\ yr}
$$

$$
\frac{A_{MedSea}*(H-H_0) + EP*A_{MedSea}*t}{3153600\ s} = \text{Supply}
$$

```python
Amed = 2500000 * 1e3 * 1e3 # m2
Hmed0 = 1500
t = 1 # year
Hmed = 1500 + 1e-3 # same depth plus 1 mm
EP = 0.8
supply = (Amed * (Hmed - Hmed0) + EP * Amed * t)/3153600
supply_sv = supply/1e6
supp = "%.4f" % supply_sv
stri = 'Average Inflow to Mostly Balance E-P and Result in 1 mm/yr SLR = ' + supp + ' Sv'
print(stri)
```

### 4.3) The narrowest point of the [Gibralter Strait](https://en.wikipedia.org/wiki/Strait_of_Gibraltar) is 14.2 km, where it is about 900 m deep. Assuming the net supply of water that balances the E-P deficit comes through that Strait, what is the average inward velocity required in cm s$^{-1}$?

```python
x = 14.2e3
z = 900
transport = supply # m3/s
u = transport / x / z
u_cm = u * 100
velo = "%.4f" % u_cm
stri = 'Average Inward Velocity Gibraltar = ' + velo + ' cm/s'
print(stri)
```

```python
# Define a reference density, which I took as
# the mean density over depth in 2015-2022
rho0 = ds15['rho'].squeeze().mean()

# solve for 1/rho * drho/dt
dt = 60 # years
drhodt = (ds55['rho'].squeeze() - ds15['rho'].squeeze())/dt/rho0

# drhodt is the density difference for each depth level
# so multiply each depth level by the thickness (derived from depth_bnds)
# and sum over all depth levels
drhodtdz_int = (drhodt * ds15['depth_bnds'].diff('nbounds').squeeze()).sum()

print('Steric sea level rise (no thermal forcing or movement) from 1955-1964 to 2015-2022 = ' + str(round(float(drhodtdz_int), 10)) + ' m')
print('Steric sea level rise (no thermal forcing or movement) from 1955-1964 to 2015-2022 = ' + str(round(float(drhodtdz_int/1e-3), 10)) + ' mm')
```
